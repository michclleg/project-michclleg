```python
import time
import os
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
fablib = fablib_manager(project_id="f5b4fc7d-978f-45be-9530-b38db8ef5046")

slice_name = "libcompart"
image = "default_ubuntu_22"
site = "NCSA"

slice = fablib.new_slice(name = slice_name)
```

```python
#=============  Node Creation  ===============================================================================#
for name in ["n1", "n2", "n3", "n4", "n5"]:
    node = slice.add_node(name = name, image = image, cores = 2, ram = 4, disk = 9, site = site)
    iface = node.add_component(model = "NIC_Basic", name = "iface1").get_interfaces()[0]
    globals()[f"{name}_iface1"] = iface

net = slice.add_l2network(name = "net", interfaces = [
    n1_iface1, n2_iface1, n3_iface1, n4_iface1, n5_iface1
])

slice.submit()
```

```python
#=============  Validate Slice  ==============================================================================#
print("=== Validating slice ===")
all_ok = True
for node in slice.get_nodes():
    comps = node.get_components()
    l2_ifaces = [
        i for c in comps for i in c.get_interfaces()
        if i.get_network() and i.get_network().get_name() == "net"
    ]
    os_name = l2_ifaces[0].get_physical_os_interface_name() if l2_ifaces else "MISSING"
    status = "OK" if l2_ifaces else "BROKEN"
    print(f"  {node.get_name()}: {status} -> {os_name}")
    if not l2_ifaces:
        all_ok = False

if not all_ok:
    print("\nERROR: Slice is broken — delete and re-submit before continuing.")
else:
    print("\nSlice is healthy")
    
slice = fablib.get_slice(slice_name)
```

```python
#=============  Install GCC and GMP on all nodes  ============================================================#
install_cmd = "sudo apt-get update -qq && sudo apt-get install -y gcc libgmp-dev"

for node in slice.get_nodes():
    print(f"Installing dependencies on {node.get_name()}...")
    node.execute(install_cmd)
    print(f"  {node.get_name()}: done")
```

```python
#=============  Upload worker and coordinator files  =========================================================#
for node in slice.get_nodes():
    print(f"Uploading worker.c on {node.get_name()}...")
    node.upload_file('worker.c', 'worker.c')
    print(f"  {node.get_name()}: done")

n1_node = slice.get_node('n1')
print("Uploading coordinator.c on n1...")
n1_node.upload_file('coordinator.c', 'coordinator.c')
print("  n1: done")

#=============  Compile workers and coordinator  =============================================================#
compile_worker = "gcc -O2 -o worker worker.c -lgmp -lm"
for node in slice.get_nodes():
    node.execute(compile_worker)
    print(f"Compile worker on {node.get_name()}: done")
    
compile_coord = "gcc -O2 -o coordinator coordinator.c -lgmp -lm -lpthread"
n1_node.execute(compile_coord)
print("Compile coordinator on n1: done")
```

```python
#=============  Retrieve node handles  =======================================================================#
n1_node = slice.get_node("n1")
n2_node = slice.get_node("n2")

ip_assignments = {
    "n1": "192.168.1.1",
    "n2": "192.168.1.2",
    "n3": "192.168.1.3",
    "n4": "192.168.1.4",
    "n5": "192.168.1.5",
}

#=============  Configure L2 interfaces  =====================================================================#
print("=== Configuring L2 interfaces ===")
for node in slice.get_nodes():
    name = node.get_name()
    ip = ip_assignments[name]
    l2_iface = next(
        i for c in node.get_components()
        for i in c.get_interfaces()
        if i.get_network() and i.get_network().get_name() == "net"
    )
    os_iface = l2_iface.get_physical_os_interface_name()
    stdout, stderr = node.execute(
        f"sudo ip addr flush dev {os_iface} && "
        f"sudo ip addr add {ip}/24 dev {os_iface} && "
        f"sudo ip link set {os_iface} up"
    )
    print(f"  {name}: {ip} on {os_iface} -> {'OK' if not stderr.strip() else stderr.strip()}")
time.sleep(2)

#=============  Verify connectivity  =========================================================================#
print("\n=== Testing L2 connectivity from n1 ===")
for name, ip in ip_assignments.items():
    if name == "n1":
        continue
    stdout, _ = n1_node.execute(f"ping -c 2 -W 2 {ip} 2>&1 | tail -2")
    print(f"  n1 -> {name} ({ip}): {stdout.strip()}")

#=============  Stop existing workers  =======================================================================#
print("\n=== Stopping existing workers ===")
for node in slice.get_nodes():
    node.execute("pkill -f worker 2>/dev/null || true")
    print(f"  {node.get_name()}: stopped")
time.sleep(1)

#=============  Start workers  ===============================================================================#
print("\n=== Starting workers ===")
for node in slice.get_nodes():
    name = node.get_name()
    for port in [5000, 5001]:
        node.execute(f"nohup ./worker {port} > worker_{port}.log 2>&1 &")
    print(f"  {name}: workers started on ports 5000, 5001")
time.sleep(2)

#=============  Verify workers are listening  ================================================================#
print("\n=== Verifying workers ===")
for node in slice.get_nodes():
    stdout, _ = node.execute("ss -tlnp | grep -E '5000|5001'")
    print(f"  {node.get_name()}: {stdout.strip()}")
print("Done")
```
